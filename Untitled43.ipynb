```python
import time
import matplotlib.pyplot as plt
import random

def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

def fib_dp(n):
    if n <= 1:
        return n
    dp = [0]*(n+1)
    dp[1] = 1
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]
fib_sizes = [5, 10, 15, 20, 25, 30]
fib_recursive_times = []
fib_dp_times = []

for n in fib_sizes:
    start = time.time()
    fib_recursive(n)
    fib_recursive_times.append(time.time() - start)

    start = time.time()
    fib_dp(n)
    fib_dp_times.append(time.time() - start)
def bubble_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def insertion_sort(arr):
    arr = arr.copy()
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >=0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

def selection_sort(arr):
    arr = arr.copy()
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr)//2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    merged = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1
    merged += left[i:]
    merged += right[j:]
    return merged

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[0]
    left = [x for x in arr[1:] if x < pivot]
    right = [x for x in arr[1:] if x >= pivot]
    return quick_sort(left) + [pivot] + quick_sort(right)

sort_sizes = [100, 200, 300, 400, 500]
bubble_times = []
insertion_times = []
selection_times = []
merge_times = []
quick_times = []

for size in sort_sizes:
    arr = [random.randint(0,1000) for _ in range(size)]

    start = time.time()
    bubble_sort(arr)
    bubble_times.append(time.time() - start)

    start = time.time()
    insertion_sort(arr)
    insertion_times.append(time.time() - start)

    start = time.time()
    selection_sort(arr)
    selection_times.append(time.time() - start)

    start = time.time()
    merge_sort(arr)
    merge_times.append(time.time() - start)

    start = time.time()
    quick_sort(arr)
    quick_times.append(time.time() - start)

def linear_search(arr, target):
    for i in arr:
        if i == target:
            return True
    return False

def binary_search(arr, target):
    arr = sorted(arr)
    left, right = 0, len(arr)-1
    while left <= right:
        mid = (left+right)//2
        if arr[mid] == target:
            return True
        elif arr[mid] < target:
            left = mid+1
        else:
            right = mid-1
    return False
search_sizes = [1000, 2000, 3000, 4000, 5000]
linear_times = []
binary_times = []

for size in search_sizes:
    arr = [random.randint(0,10000) for _ in range(size)]
    target = random.choice(arr)

    start = time.time()
    linear_search(arr, target)
    linear_times.append(time.time() - start)

    start = time.time()
    binary_search(arr, target)
    binary_times.append(time.time() - start)

plt.figure(figsize=(6,4))
plt.plot(fib_sizes, fib_recursive_times, marker='o')
plt.title("Fibonacci Recursive")
plt.xlabel("n")
plt.ylabel("Time (s)")
plt.show()

plt.figure(figsize=(6,4))
plt.plot(fib_sizes, fib_dp_times, marker='o', color='orange')
plt.title("Fibonacci DP")
plt.xlabel("n")
plt.ylabel("Time (s)")
plt.show()

def plot_individual_sort(times, sizes, name):
    plt.figure(figsize=(6,4))
    plt.plot(sizes, times, marker='o')
    plt.title(name)
    plt.xlabel("Input Size")
    plt.ylabel("Time (s)")
    plt.show()

plot_individual_sort(bubble_times, sort_sizes, "Bubble Sort")
plot_individual_sort(insertion_times, sort_sizes, "Insertion Sort")
plot_individual_sort(selection_times, sort_sizes, "Selection Sort")
plot_individual_sort(merge_times, sort_sizes, "Merge Sort")
plot_individual_sort(quick_times, sort_sizes, "Quick Sort")

plt.figure(figsize=(6,4))
plt.plot(search_sizes, linear_times, marker='o', label="Linear Search")
plt.title("Linear Search")
plt.xlabel("Input Size")
plt.ylabel("Time (s)")
plt.show()

plt.figure(figsize=(6,4))
plt.plot(search_sizes, binary_times, marker='o', label="Binary Search", color='green')
plt.title("Binary Search")
plt.xlabel("Input Size")
plt.ylabel("Time (s)")
plt.show()


plt.figure(figsize=(6,4))
plt.plot(fib_sizes, fib_recursive_times, marker='o', label="Recursive")
plt.plot(fib_sizes, fib_dp_times, marker='o', label="DP")
plt.title("Fibonacci: Recursive vs DP")
plt.xlabel("n")
plt.ylabel("Time (s)")
plt.legend()
plt.show()
plt.figure(figsize=(7,5))
plt.plot(sort_sizes, bubble_times, marker='o', label="Bubble Sort")
plt.plot(sort_sizes, insertion_times, marker='o', label="Insertion Sort")
plt.plot(sort_sizes, selection_times, marker='o', label="Selection Sort")
plt.plot(sort_sizes, merge_times, marker='o', label="Merge Sort")
plt.plot(sort_sizes, quick_times, marker='o', label="Quick Sort")
plt.title("Sorting Algorithms Comparison")
plt.xlabel("Input Size")
plt.ylabel("Time (s)")
plt.legend()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(search_sizes, linear_times, marker='o', label="Linear Search")
plt.plot(search_sizes, binary_times, marker='o', label="Binary Search")
plt.title("Searching Algorithms Comparison")
plt.xlabel("Input Size")
plt.ylabel("Time (s)")
plt.legend()
plt.show()
```
